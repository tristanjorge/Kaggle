```python
# use plotly instead of geopandas for better visualization, interaction with plots

# import plotly libraries
import plotly.express as px
import plotly.graph_objs as go
from plotly.subplots import make_subplots

# user-defined plot_map function for easier plotting
def plot_worldmap(df, col, pal):
    fig = px.choropleth(
        df, locations="Country/Region", locationmode='country names',
        color=col, 
        hover_name="Country/Region",
        title=f"World Map ({col})", 
        hover_data=[col], 
        color_continuous_scale=pal,
        width = 1024,
        height = 768
        )

    fig.show()
```

```python
# Plotting world map with "Confirmed"
plot_worldmap(country_wise, 'Confirmed', 'matter')
```

```python
# Plotting world map with Deaths
plot_worldmap(country_wise, 'Deaths', 'matter')
```

```python
# Plotting world map with Deaths / 100 cases
plot_worldmap(country_wise, 'Deaths / 100 Cases', 'matter')
```

```python
# Cases over time from 22 Jan 2020 - 27 Jul 2020

fig = px.choropleth(
    full_grouped, 
    locations = "Country/Region",
    color = np.log(full_grouped["Confirmed"]),
    locationmode = "country names",
    hover_name = "Country/Region",
    animation_frame = full_grouped["Date"].dt.strftime('%Y-%m-%d'),
    title = "Covid 19 cases from 22 Jan 2020 to 27 Jul 2020",
    color_continuous_scale = px.colors.sequential.matter,
    width = 1024,
    height = 768
    )

fig.show()
```

```python
# bar chart analysis of top contributing countries

# user defined function for easier plotting, using worldometer data
def plot_hbar(col, n, hover_data=[]):
    df = worldometer_data
    fig = px.bar(
        df.sort_values(col).tail(n),
        x = col, y = "Country/Region",
        color = "WHO Region",
        text = col, 
        width = 1024,
        height = 768,
        color_discrete_sequence = px.colors.qualitative.Bold,
        title = f"{col} (Worldometer)"
    )
    fig.update_layout(
        xaxis_title ="Cases", yaxis_title="Country/Region",
        yaxis_categoryorder = 'total ascending',
        uniformtext_minsize=8
    )

    fig.update_traces(
        texttemplate='%{text:,}'
    )

    fig.show()
```

```python
# worldometer_data info
worldometer_data.info()
```

```python
# Bar chart - Total Cases
plot_hbar('TotalCases', 20)
```

```python
# Bar chart - Total Deaths
plot_hbar('TotalDeaths', 20)
```

```python
# Bar chart - Total Recovered
plot_hbar('TotalRecovered', 20)
```

```python
# for 1M pop data, need to exclude countries / regions with less than 1M population
# do this using a user-defined function, with worldometer data
def plot_hbar_1m(col, n):
    df = worldometer_data[worldometer_data['Population']>1e6]
    fig = px.bar(
        df.sort_values(col).tail(n),
        x = col, y = "Country/Region",
        color = "WHO Region",
        text = col,
        width = 1024,
        height = 768,
        color_discrete_sequence = px.colors.qualitative.Pastel,
        title = f"{col} (Countries with >1M Pop)"
    )

    fig.update_layout(
        xaxis_title ="Cases", yaxis_title="Country/Region",
        yaxis_categoryorder = 'total ascending',
        uniformtext_minsize=8
    )

    fig.update_traces(
        texttemplate='%{text:,}'
    )

    fig.show()
```

```python
# Bar chart - Total Cases / 1M Population
plot_hbar_1m('Tot Cases/1M pop', 20)
```

```python
# Bar chart - Deaths / 1M Population
plot_hbar_1m('Deaths/1M pop', 20)
```

```python
# Bar chart - Percentage of total population
worldometer_data['% Cases of Pop'] = round((worldometer_data['TotalCases'] / worldometer_data['Population']) * 100,2)

plot_hbar_1m('% Cases of Pop', 20)
```
